```python
import pandas as pd
import numpy as np
import missingno as msno
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Dataset/Uncleaned_netflix_titles.csv")
```

```python
#removing rows with missing dates
df.dropna(subset='date_added',inplace=True)
```

```python
df.info()
```

```python
#replacing null values of director field with the string 'missing'
df['director'].fillna('missing',inplace=True)
df['cast'].fillna('missing',inplace=True)
df['country'].fillna('missing',inplace=True)
df['rating'].fillna('missing',inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
cast_sep=[]
for i in range(len(df)):
    cast_sep.extend(df.cast.iloc[i].split(","))
```

```python
#print(cast_sep)
```

```python
cast_count = {}
for i in cast_sep:
    cast_count[i]=cast_sep.count(i)
#cast_count
```

```python
cast_count = sorted(cast_count.items(), key=lambda item: item[1], reverse=True)
```

```python
#cast_count=dict(cast_count) *keeping it as list for visualization purpose*
```

```python
genre_tag = []
for i in range(len(df)):
    genre_tag.extend(df.listed_in.iloc[i].split(","))
```

```python
genre_count = {}
for i in genre_tag:
    genre_count[i]=genre_tag.count(i)
```

```python
genre_count = sorted(genre_count.items(), key=lambda item:item[1], reverse=True )
```

```python
genre_count=dict(genre_count)
```

```python
df["added_year"]=np.nan
for i in range(len(df)):
    df["added_year"][i] = df.date_added.iloc[i].split(",")[1]
```

```python
df["added_month"]=np.nan
for i in range(len(df)):
    df["added_month"][i] = df.date_added.iloc[i].split(" ")[0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# text annotation for vertical bar chart
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]/2, y[i], ha = 'center')
```

```python
plt.style.use('default')
plt.figure(figsize=(3,3))
plt.bar(df.type.unique(), df.type.value_counts(), width=0.5, color='red',edgecolor='black')
addlabels(df.type.unique(), df.type.value_counts())
plt.ylabel('Value Count')
plt.show()
```

From the above graph, its clear that Netflix added more number of TV Shows than movies between the year 2008-2021

```python
#Function to add center text annotations in horizontal barchart
def add_text(values):
    for i,v in enumerate(values):
        plt.text(v/2,i,str(v),color='black',fontweight='normal',ha='center',va='center')
```

```python
#Top 10 Genres(tags)
plt.figure(figsize=(7,4))
top_genre = dict(list(genre_count.items())[:10])

# Create lists of keys and values
keys = list(top_genre.keys())[::-1]
values = list(top_genre.values())[::-1]

# Create a horizontal bar chart
plt.barh(keys, values, color='red',edgecolor='black')
add_text(values)
plt.xlabel('Value Count')
plt.title('Most Used Tag')
plt.show()
```

International movies is the most common tag used in Netflix

```python
#Most Common Cast
plt.figure(figsize=(7,4))
dict_cast = dict(cast_count[1:])
top_cast = dict(list(dict_cast.items())[:10])
keys = list(top_cast.keys())[::-1]
values = list(top_cast.values())[::-1]
plt.barh(keys,values,color='red',edgecolor='black')
#add text annotations
add_text(values)
plt.xlabel('Value Count')
plt.title('Most Common Cast')
plt.show()
```

Anupam Kher is the most common cast found over Netflix (This dataset has over 700 missing values in cast)

```python
country_sep=[]
for i in range(len(df.country)):
    country_sep.extend(df.country.iloc[i].split(","))

country_count={}
for i in country_sep:
    country_count[i]=country_sep.count(i)
```

```python
country_count = dict(sorted(country_count.items(), key= lambda item: item[1], reverse=True))
```

```python
#Top 15 Countries
plt.figure(figsize=(12,6))
top_country=dict(list(country_count.items())[:15])
keys=list(top_country.keys())[::-1]
values=list(top_country.values())[::-1]
plt.barh(keys,values,color='red',edgecolor='black',height=0.7)
add_text(values)
plt.xlabel('Value Count')
plt.title('Top 15 Countries')
plt.show()
```

Most number of TV Shows and Movies were given by United States

```python
#Release Year frequency for each type
df_movie = pd.DataFrame(df.groupby('type')['release_year'].value_counts()['Movie'].sort_values(ascending=False)[:20])
df_tv = pd.DataFrame(df.groupby('type')['release_year'].value_counts()['TV Show'].sort_values(ascending=False)[:20])

df_stacked = pd.concat([df_movie, df_tv], axis=1)
df_stacked.columns = ['Movies', 'TV Shows']

df_stacked.plot(kind='bar', stacked=True)
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Top 20 Movies and TV Shows by Release Year')
plt.show()
```

2017 and 2018 had most amount of movies released, where as maximum number of TV shows were released in 2020.
